```python
import pandas as pd

#HELIOS
url="https://cdsarc.cds.unistra.fr/viz-bin/nph-Cat/csv?-plus=-+&J/A%2BA/623/A85/isochrones/BEX%5Fevol%5Fmags%5F-3%5FMH%5F0.00.dat"
#petitCODE
# url="https://cdsarc.cds.unistra.fr/viz-bin/nph-Cat/csv?-plus=-+&J/A%2BA/623/A85/isochrones/BEX%5Fevol%5Fmags%5F-2%5FMH%5F0.00.dat"
df = pd.read_csv(url,comment="#", delimiter="\t", header=[0,1])
df
```

```python
from astropy import units as u 

df = df.iloc[1:]
df = df.droplevel(level=1, axis=1).astype(float)

df["Me"] = df["Mass"].copy()
df["Rj"] = df["Rad"].copy()
df["Mj"] = df["Me"]*u.Mearth.to(u.Mjup)
df["Re"] = df["Rj"]*u.Rjup.to(u.Rearth)
df
```

```python
planets = {
    "K2-100 b": [[3.88,0.16], [21.8,6.2]], #Baraggan+20
    "TOI-560 b": [[2.79,0.10],[10.2,3.4]], #Baraggan+22
    "TOI-560 c": [[2.39,0.10],[9.7,1.8]], #Baraggan+22; El Mufti+23
    "TOI-179 b": [[2.62,0.1],[24,7]], #Desidera+23
    "K2-136 c": [[3.00,0.13],[18.1,1.9]], #Mayo+23
    "TOI-1726 c": [[2.692,0.108],[15.54,3.86]], #Mallorquin+23
    "AUMic b": [[4.38,0.18],[20.12,1.7]], #Cale+21; Mp=17.1±4.7 from Klein+21; Mp=11.7±5.0 from Zicher+22
    "AUMic c": [[3.51,0.16],[22.2, 6.7]], #Zicher+2021
    "TOI-1807 b": [[1.37,0.09],[2.57,0.50]], #Nardiello+22 
    "K2-25 b": [[3.44,0.12],[24.5,5.7]], #stefansson+22
    "TOI-1201 b": [[2.415, 0.090],[6.28, 0.88]], #Kossakowski+2023
    "TOI-1099 b": [[2.25,0.12],[6.1,1.6]], #Barros+2023
#     "TOI-1268 b": [[9.1, 0.6], [96.4, 8.3]], #Subjak+23
#     "TOI-4562 b": [[12.53,0.1569],[731.005, 152.557]], #Heitzmann+23 Rj=1.118−0.014+0.013, Mj=2.30+0.48 
}
# gas giants
planets2 = {
    "TOI-4562 b": [[1.118,0.014],[2.30,0.48]], #Heitzmann+23 Rj=1.118−0.014+0.013, Mj=2.30+0.48 
    "TOI-1268 b": [[0.811, 0.0535], [0.303, 0.0535]], #Subjak+23
}
```

```python
import matplotlib.pyplot as pl

fig, ax = pl.subplots()

for p in planets:
    d = planets[p]
    r,m = d[0], d[1]
    ax.errorbar(m[0], r[0], xerr=m[1], yerr=r[1])
    ax.annotate(xy=[m[0], r[0]], text=p)
```

```python
import matplotlib.pyplot as pl

fig, ax = pl.subplots()

g = df.groupby("logAge")
for i in g.groups:
    if float(i) in [6,7,8, 9]:
        d = g.get_group(i)
        d.plot(x="Me",
               y="Re",
               lw=2,
               #marker='o',
               ax=ax,
               label=10**(float(i)-6)
              )
ax.set_xlim(0,100)
ax.legend(title="Age (Myr)")
ax.set_ylabel("Planet Radius ("+r"R$_{\oplus}$)")
ax.set_xlabel("Planet Mass ("+r"M$_{\oplus}$)")

for p in planets:
    d = planets[p]
    r,m = d[0], d[1]
    ax.errorbar(m[0], r[0], xerr=m[1], yerr=r[1], c='k')
    ax.annotate(xy=[m[0], r[0]], text=p)
```

```python
import matplotlib.pyplot as pl

fig, ax = pl.subplots()

g = df.groupby("logAge")
for i in g.groups:
    if float(i) in [6,7,8, 9]:
        d = g.get_group(i)
        d.plot(x="Mj",
               y="Rj",
               lw=2,
               #marker='o',
               ax=ax,
               label=10**(float(i)-6)
              )
ax.legend(title="Age (Myr)")
ax.set_ylabel("Planet Radius ("+r"R$_{\rm{Jup}}$)")
ax.set_xlabel("Planet Mass ("+r"M$_{\rm{Jup}}$)")

for p in planets2:
    d = planets2[p]
    r,m = d[0], d[1]
    ax.errorbar(m[0], r[0], xerr=m[1], yerr=r[1], c='k')
    ax.annotate(xy=[m[0], r[0]], text=p)
ax.set_xscale("log")
ax.set_xlim(-1,5)
```
