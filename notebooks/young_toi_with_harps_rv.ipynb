# NExSci

```python
import sys
sys.path.insert(0, '../code/')
from catalog import get_nexsci_data

nexsci = get_nexsci_data(table_name="ps", clobber=False)
nexsci.head()
```

```python
idx1 = nexsci.hostname.str.lower().replace('([- ])', '', regex=True).isin(df2.Target.str.lower().replace('([- ])', '', regex=True))
idx2 = nexsci.hd_name.str.lower().replace('([- ])', '', regex=True).isin(df2.Target.str.lower().replace('([- ])', '', regex=True))
idx3 = nexsci.hip_name.str.lower().replace('([- ])', '', regex=True).isin(df2.Target.str.lower().replace('([- ])', '', regex=True))
nexsci[idx1|idx2|idx3]
```

```python
idx1.sum(), idx2.sum(), idx3.sum()
```

```python
hostnames = nexsci[idx1|idx2|idx3][['hostname','hip_name','hd_name']].stack().str.lower().replace('([- ])', '', regex=True).unique()
len(hostnames)
```

```python
idx=df2.Target.str.lower().replace('([- ])', '', regex=True).isin(hostnames)
df2[idx].shape
```

```python
from astropy.io import ascii
from tqdm import tqdm

data = {}
base_url = 'https://www2.mpia-hd.mpg.de/homes/trifonov/Ver_02/'
for i,d in tqdm(df2[idx].iterrows()):
    fn = d['Data products (csv)']
    name = fn.split('_')[0]
    url = base_url+name+'_RVs/'+fn
    try:
        rv = ascii.read(url).to_pandas()
        data[name] = len(rv)
    except Exception as e:
        print(e)
```

```python
pd.Series(data).sort_values()
```
