## Young in nexsci

```python
nexsci_young = nexsci[nexsci.st_age<2]
```

```python
from astropy.coordinates import SkyCoord

d = df2.drop_duplicates(subset=['target'], keep='first').dropna(subset=['ra','dec'])
c1 = SkyCoord(d.ra.values, d.dec.values, unit='degree')
```

```python
c2 = SkyCoord(nexsci_young.ra.values, nexsci_young.dec.values, unit='degree')
```

```python
idx, d2d, _ = c1.match_to_catalog_sky(c2)
```

```python
import astropy.units as u

max_sep = 1.0 * u.arcsec
sep_constraint = d2d < max_sep

# Filter matched rows
cols1 = ['target']
matched_df1 = d[sep_constraint][cols1].reset_index(drop=True)
cols2 = ['hostname']
matched_df2 = nexsci_young.iloc[idx[sep_constraint]][cols2].reset_index(drop=True)

counts = df2['target'].value_counts()
matched_df1['N_RVs'] = matched_df1['target'].map(counts).fillna(0).astype(int)
matched_df = pd.concat([matched_df1, matched_df2], axis=1, keys=['df1', 'df2'])
```
