## TOI catalog

```python
from astropy.table import Table

fp = "../data/kepler_dr3_good.fits"
kepler = Table.read(fp, format='fits')
df_kep = kepler.to_pandas()
df_kep.head()
```

```python
fp = "../data/k2_dr2_1arcsec.fits"
k2 = Table.read(fp, format='fits')
df_k2 = k2.to_pandas()
df_k2.head()
```

```python
fp = "../data/exoplanets_dr2_1arcsec.fits"
nexsci = Table.read(fp, format='fits')
df_nexsci = nexsci.to_pandas()
df_nexsci.head()
```

## cluster catalog

```python
import pandas as pd

#https://ucc.ar/faq/#where-is-the-full-database
df = pd.read_parquet("../data/UCC_members.parquet.gz")
df.shape
```

```python
idx=df.Source.isin(df_kep.source_id)
idx.sum()
```

```python
ids = df[idx][df[idx].probs>0.99].Source.values
idx2=df_kep.source_id.isin(ids)
kep_cluster_mem = df_kep[idx2]
kep_cluster_mem['planet?'].unique()
```
