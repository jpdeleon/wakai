```python
from catalog import CatalogDownloader

cd = CatalogDownloader(catalog_name="Hunt2024", catalog_type="cluster")
tabs = cd.get_tables_from_vizier(row_limit=-1, save=True)
```

```python
df = tabs[0].to_pandas()
df_mem = tabs[1].to_pandas()
df_mem.head()
```

```python
#replace name to be compatible with other catalogs' naming formats
df_mem = df_mem.rename({#"Cluster": "Name",
                        # "Source": "GaiaDR3",
                        # "Proba": "Prob",
                        "RA_ICRS": "ra",
                        "DE_ICRS": "dec",
                        "pmRA": "pmra",
                        "pmDE": "pmdec",
                        "Plx": "parallax",
                        "RV": "radial_velocity"
                        }, axis=1)
df_mem.head()
```

```python
from catalog import get_nexsci_data

nexsci = get_nexsci_data(clobber=False)
nexsci.tail()
```

# cross-match nexsci with catalog

```python
import pandas as pd

if True:
    #from astroquery.xmatch
    matches = pd.read_csv('../data/nexsci_GaiaDR3_xmatch.csv')
if False:
    #from vizier query/exofop; not yet available
    matches = pd.read_csv('../data/nexsci_with_Gaiaid.csv')
    matches['GaiaDR3'] = matches['GaiaDR3_vizier']
matches.head()
```

```python
final = pd.merge(df_mem, matches, on='GaiaDR3')
final.head()
```

```python
final.hostname.unique().shape
```

```python
final_default = final[final['default_flag']==1]
final_default.shape
```

```python
final_default.sort_values(by='pl_name')\
    .to_csv(f'../data/{cd.catalog_name}/nexsci_in_{cd.catalog_name}.csv', 
            index=False)
```
